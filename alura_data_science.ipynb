```python
import pandas as pd

notas = pd.read_csv("/content/drive/MyDrive/pasta para Alura's file/ratings.csv")
print(notas["rating"].unique(),end='\n\n')
print(notas['rating'].value_counts(), end='\n\n')
print(notas.userId.value_counts(), end='\n\n')
print(notas.rating.plot(kind='hist'))
print(notas.rating.median())
notas.head()
```

```python
movies = pd.read_csv("/content/drive/MyDrive/pasta para Alura's file/tmdb_5000_movies.csv")
print(movies.original_language.unique(), end='\n\n')
movies.original_language.value_counts().to_frame()   # transformar em DataFrame
```
